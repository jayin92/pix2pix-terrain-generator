```python
path = "E:/scifair/heightmap/China256/150,450_China.png"
size = 256
ori_im = cv2.imread(path)
im = ori_im
r = 64
im = ori_im
im = cv2.resize(im, (size//r, size//r))
im = cv2.resize(im, (size, size))

plt.imshow(im)
cv2.imwrite("r{}wm.png".format(r), im)
im = cv2.medianBlur(im, 31)
cv2.imwrite("r{}.png".format(r), im)
```
